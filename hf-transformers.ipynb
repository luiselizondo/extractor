```python
import os
from transformers import pipeline, AutoTokenizer, AutoConfig, AutoModelForPreTraining, AutoModelForDocumentQuestionAnswering, AutoModelForSequenceClassification
```

```python
dir_cur = os.getcwd()
pdf_file = "nike-paper.mmd"
input_file = os.path.join(f"{dir_cur}/output/", pdf_file)
```

```python
content = open(input_file, "r").read()
content_lines = open(input_file, "r").readlines()
```

```python
prompt = """
  Given one question and its corresponding answer based on the last context in the last
  example. Follow the format of the examples below to include context, question, and answer in the response.
  """

model = AutoModelForPreTraining.from_pretrained("gpt2")
tokenizer = AutoTokenizer.from_pretrained("gpt2")

generator = pipeline("text-generation", model=model, tokenizer=tokenizer)
encoding = tokenizer(content_lines)

results = generator(prompt, num_return_sequences=2, temperature=0.7)
```

```python
for result in results:
    print(result["generated_text"])

print(encoding)
```
